```python
from datetime import datetime, date
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
def gettimestamp(vdate):
    time_obj = datetime.strptime(vdate,'%y %m %d %H %M %S.%f')
    timestamp = time_obj.timestamp()
    return timestamp
```

```python
data = pd.read_csv('csv_files/211011117.17N.csv')
print(data.columns)
d = dict()
prn = data['prn']
epoch_time = data['epoch_time']
param = data['fit_interval']
for i in range(len(prn)):
    if prn[i] not in d:
        d[prn[i]]=list()
    d[prn[i]].append([gettimestamp(epoch_time[i]),param[i]])    
```

```python
for prn,values in d.items():
    print('Satellite PRN : ',prn)
    for val in values:
        plt.figure(len(d.items()),figsize=(30,5))
        plt.subplot(222)
        plt.plot(val[:][0:1],val[:][1:2],'ro')
    plt.show()
```
